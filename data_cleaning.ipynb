```python
def answer_key(path):
    test1=pd.read_csv(path, sep='delimiter', header=None)
    for t_num in test1.iloc[2]:
        test_num = []
        data=t_num.split(",")
        num=data[1]
        test_num.append(num)
        test_num=pd.DataFrame({"test_num":test_num})

    for ans in test1.iloc[4]:
        data=ans.split(",")
        test_ans=pd.DataFrame({"ans": data[3:]}).T.reset_index(drop="index")
    
    test_ans=pd.concat([test_num, test_ans], axis=1).drop(154, axis=1)
    return test_ans
```
