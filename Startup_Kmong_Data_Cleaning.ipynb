```python
# Write your code here!
os =pd.DataFrame(raw_log['osversion'], columns =['osversion'])
```

```python
os['osversion'].unique()
```

```python
# fill in ostype(clean) with ios
os['ostype(clean)'] = 'iOS'
os['osversion(clean)'] = '11.4.1'
```

```python
# insert Android if it has Android
os.loc[os['osversion'].str.contains("Android"), 'ostype(clean)'] = 'Android'
```

```python
# get all android osversion
android = []
for name in os.loc[os['osversion'].str.contains("Android"), 'osversion']:
    android.append(name.split("Android")[1])
# get all iOS osversion
ios = []
for name in os.loc[os['osversion'].str.contains("iOS"), 'osversion']:
    ios.append(name.split("iOS")[1]) 
```

```python
# insert Android Os Version 
os.loc[os['osversion'].str.contains("Android"), 'osversion(clean)'] = android
```

```python
# insert Android Os Version 
os.loc[os['osversion'].str.contains("iOS"), 'osversion(clean)'] = ios
```

```python
df_channel = pd.DataFrame(raw_log['channel'])
df_channel.info()
```

```python
# there are null values: Null -> No
df_channel.loc[df_channel['channel'].isnull()] = "no"
```

```python
df_channel.loc[df_channel['channel'].str.contains("google")] = "google"
df_channel.loc[df_channel['channel'].str.contains("m_daum")] = "daum"
df_channel.loc[df_channel['channel'].str.contains("m_naver|pc_naver|m_naverpowercontents")] = "naver"
df_channel.loc[df_channel['channel'].str.contains("apple.searchads")] = "apple"
df_channel.loc[df_channel['channel'].str.contains("WEB")] = 'web'
```

```python
# fill nan back 
import numpy as np
df_channel.loc[df_channel['channel'].str.contains('no')] = np.nan
```

```python
# check the answer 
df_channel['channel'].value_counts()
```

```python
df_view = pd.DataFrame(raw_log['inappeventcategory'])
```

```python
# null -> na
df_view.loc[df_view['inappeventcategory'].isnull()] = "no"
```

```python
# creating view id and view action first
viewid = []
for name in df_view['inappeventcategory']:
    viewid.append(name.split("."))
```

```python
viewid = pd.DataFrame(viewid, columns=['viewid', 'viewaction'])
```

```python
# make viewcategory
viewcate = []
for name in viewid['viewid']:
    viewcate.append(name.split("_")[0])
```

```python
df_view['viewcategory'] = viewcate
df_view["viewid"] = viewid['viewid']
df_view['viewcation'] = viewid['viewaction']
```

```python
# back to null 
df_view.loc[df_view['inappeventcategory'].str.contains("no")] = np.nan
```
