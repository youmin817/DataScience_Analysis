### categorical plots !! comparing !! 

```python
df_cate = pd.read_csv('../data/laptops2.csv')
```

```python
sns.catplot(data= df_cate, x='os', y='price', kind='box')
```

```python
# box plot doesn't help when data is small!!
# use strip to see all data 
sns.catplot(data= df_cate, x='os', y='price', kind='strip')
```

```python
# See the plot by Processor_brand
# use hue to see them 
# hue = color !! So By processor_brand, it gives different colors 
sns.catplot(data= df_cate, x='os', y='price', kind='strip', hue = "processor_brand")
```

```python
# use swarm plot to see more detail !!
sns.catplot(data= df_cate, x='os', y='price', kind='swarm', hue = "processor_brand")
```
