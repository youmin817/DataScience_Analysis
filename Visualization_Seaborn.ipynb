```python
import pandas as pd
```

```python
url= "https://bit.ly/dsa-04-health"

data = pd.read_csv(url)
print(data.shape)
data.head()
```

### Explore

```python
# defalut option
import seaborn as sns
import matplotlib # option
from IPython.display import set_matplotlib_formats

matplotlib.rc('font', family ="AppleGothic") # For Korean
# Retina
set_matplotlib_formats('retina')

# minus
matplotlib.rc('axes', unicode_minus=False)
```

```python
sns.countplot(data=data, x="성별코드")
```

```python
### Size
import matplotlib.pyplot as plt

plt.figure(figsize =(18,3))
sns.countplot(data=data, x="시도코드")

```

```python
# mutliple plots

figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

figure.set_size_inches(18,3)
sns.countplot(data=data, x="시도코드", ax=ax1)
sns.countplot(data=data, x="성별코드", ax=ax2)

```

```python
#color

plt.figure(figsize =(18,3))
sns.countplot(data=data, x="시도코드", color='b')

```

```python
# color mix options
palette = sns.color_palette('deep', 10) # mix 10 color with deep content
sns.palplot(palette)
```

```python
# color mix
plt.figure(figsize =(18,3))
sns.countplot(data=data, x="시도코드", palette= palette)

```

```python
# hue /  by "성별코드"
plt.figure(figsize =(18,3))
sns.countplot(data=data, x="시도코드", hue= "성별코드", palette= palette)
```

```python
# x-> y vertically plot

plt.figure(figsize =(4,9))
sns.countplot(data=data, y="시도코드", hue= "성별코드", palette= palette)
```

```python
## distplot
sns.distplot(data['허리둘레'])
```

```python
sns.distplot(data['허리둘레'], hist=False)
```

```python
sns.distplot(data['허리둘레'], kde=False)
```

```python
one=data[data['흡연상태'] == 1]
two=data[data['흡연상태'] == 2]
three=data[data['흡연상태'] == 3]

sns.distplot(one['허리둘레'], hist=False, label="None Smoker")
sns.distplot(two['허리둘레'], hist=False, label="quit")
sns.distplot(three['허리둘레'], hist=False, label="Smoker")
```

```python
import numpy as np
#bar plot
# countplot : caanot set up y axis
# bar plot is possible for y axis
plt.figure(figsize=(18,3))
sns.barplot(data=data, x='연령대코드(5세단위)', y="수축기혈압") #, estimator = np.sum)
# estimator: opertions can be applied 
```

```python
# x-> y
plt.figure(figsize=(4,9))
sns.barplot(data=data, y='연령대코드(5세단위)', x="수축기혈압", hue= "성별코드", orient='h')
```

```python
#  boxplot

plt.figure(figsize=(18,3))
sns.boxplot(data=data, x='연령대코드(5세단위)', y="수축기혈압", hue="성별코드")
```

### Violinplot

```python
plt.figure(figsize=(18,3))
sns.violinplot(data=data, x='연령대코드(5세단위)', y="수축기혈압", hue="성별코드")
```

```python
plt.figure(figsize=(18,3))
sns.violinplot(data=data, x='연령대코드(5세단위)', y="수축기혈압", hue="성별코드", split=True) # split left man right woman
```

```python
# point plot
plt.figure(figsize=(18,3))
sns.pointplot(data=data, x='연령대코드(5세단위)', y="수축기혈압", hue="성별코드", linestyles=[':',"--"])
```

```python
# scatterplot
plt.figure(figsize=(18,18))
target = data[data['식전혈당(공복혈당)'] <150]

sns.scatterplot(data=target, x='식전혈당(공복혈당)', y="수축기혈압", hue='허리둘레', size = '총콜레스테롤' , sizes=(30,550))
```
