### Define VIP Customer  

```python
col = ["Name",
"Age(clean)",
"Height(clean)",
"Initial Weight",
"Lowest Weight",
"Target Weight",
"Status"]
df_copy = df[col].copy()
df_copy.head()
```


$$
BMI = \frac{체중(kg)}{키(m) \times 키(m)}
$$



```python
# Write your code here!
#1. Weight Loss(goal)
df_copy['Weight Loss(goal)'] = df['Initial Weight'] - df['Target Weight']
```

```python
#2. Weight Loss(current)
df_copy['Weight Loss(current)'] = df['Initial Weight'] - df['Lowest Weight']
```

```python
#3. 체질량지수(BMI)
height_meter = df_copy['Height(clean)'] / 100
df_copy['BMI'] = df_copy['Initial Weight'] /(height_meter * height_meter)
```

```python
# checking
df_copy[["Name","Weight Loss(goal)", "Weight Loss(current)", "BMI"]].head()
```

```python
# Write your code here!
def invalid_finder(row):
    if (pd.isnull(row['Age(clean)'])) | (pd.isnull(row["Height(clean)"])) | (pd.isnull(row["Initial Weight"])) | (pd.isnull(row["Target Weight"])) | (pd.isnull(row["Lowest Weight"])):
        return True
    elif (row["Height(clean)"] > 200) or (row["Height(clean)"] < 140):
        return True
    elif (row['BMI'] <18.5) |( row['BMI'] >30.0):
        return True
    elif row['Weight Loss(goal)'] <0:
        return True
    else:
        return False
```

```python
df_copy['Invalid']=df_copy.apply(invalid_finder, axis ='columns')
```

```python
# 눔의 프로그램을 결제한 구매자. (Status == "completed") False
df_copy.loc[~(df['Status'] == "completed"), "Invalid"] = False
```

```python
# checking
df_copy[df_copy['Invalid'] == True].head()
```

```python
# check if total number of data = 2619 
df_copy[df_copy['Invalid'] == True].shape
```

```python
# Write your code here!
df_copy['VIP'] = False
```

```python
# all conitions must be satisfied
def VIP_finder(row):
    if (~pd.isnull(row['Weight Loss(goal)'])) & (~pd.isnull(row["Weight Loss(current)"])) & (~pd.isnull(row["BMI"])) & (row["Weight Loss(current)"] > 10) & (row['BMI'] >30.0) & (row['Weight Loss(current)'] > row['Weight Loss(goal)']) :
        return True
    else:
        return False
```

```python
df_copy['VIP'] = df_copy.apply(VIP_finder, axis ='columns')
df_copy.loc[~(df_copy['Status'] == "completed"),'VIP'] = False
```

```python
def age_group(row):
    if (row['Age(clean)'] >= 55) & (row['Age(clean)'] <= 99):
        return "55~99"
    elif (row['Age(clean)'] >= 18) & (row['Age(clean)'] <= 24):
        return "18~24"
    elif (row['Age(clean)'] >= 25) & (row['Age(clean)'] <= 35):
        return "25~35"
    elif (row['Age(clean)'] >= 36) & (row['Age(clean)'] <= 44):
        return "36~44"
    elif (row['Age(clean)'] >= 45) & (row['Age(clean)'] <= 54):
        return "45~54"
    elif (row['Age(clean)'] >= 0) & (row['Age(clean)'] <= 17):
        return "00~17"
```
