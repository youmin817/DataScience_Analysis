```python
import sqlite3
import pandas as pd
```

Load Dataset

```python
connection = sqlite3.connect("e-commerce.sqlite")
connection
```

```python
query = "Select * From 'orders'"

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
# get 3 records
query = "Select * From 'orders' Limit 3"

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = "Select * From 'orders' Order by date DESC" #DESC Latest to Oldest

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = """
    Select * 
        From 'orders' 
        Order by price DESC
        Limit 3
""" 

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = """
    Select * 
        From 'orders'
        Where price >= 500
       
""" 

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = """
    Select * 
        From 'orders'
        Where state <> 'confirmed'
       
""" 

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = """
    Select * 
        From 'orders'
        Where price between 400 and 500
       
""" 

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = """
    Select * 
        From 'orders'
        Where address In ('Seoul', 'Incheon')
       
""" 

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = """
    Select * 
        From 'orders'
        Where price >= 500 AND state = 'confirmed'       
""" 

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = """
    Select * 
        From 'orders'
        Where price >= 500 AND state = 'confirmed'  
        Order by price DESC
        Limit 3
""" 

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = """
    Select * 
        From 'orders'
        Where address LIKE '%D%'
""" 

order = pd.read_sql(query, connection, index_col = "id")

order
```

Merge

```python
query = """
    Select * 
        From 'orders'
""" 

order = pd.read_sql(query, connection, index_col = "id")

order
```

```python
query = """
    Select * 
        From 'users'
""" 

user = pd.read_sql(query, connection, index_col = "id")

user
```

```python
query = """
    Select * 
        From 'orders'
        Left Join 'users' ON 'orders'.user_id = 'users'.id  
""" 

pd.read_sql(query, connection)
```

```python
query = """
    Select 'orders'.id, 'orders'.date, 'orders'.price, 'users'.name, 'users'.gender 
        From 'orders'
        Left Join 'users' ON 'orders'.user_id = 'users'.id  
""" 

pd.read_sql(query, connection)
```
